```python
text= pd.read_csv("C:/Users/juand/Documents/Python/Datathon2019/Dataset/Reddit/RC_2017-01.csv")['body']
```

```python
text_in_chain = ''
for paragraph in text:
    if '[' not in paragraph:
        text_in_chain += paragraph        
```

```python
import nltk
```

```python
nltk.download("punkt")
```

```python
text_words = nltk.word_tokenize(text_in_chain)
```

```python
nltk.download("stopwords")
```

```python
import string
```

```python
string.punctuation
```

```python
useless_words = nltk.corpus.stopwords.words("english") + list(string.punctuation) +['"','...','``',"''"]
useless_words
```

```python
def build_bag_of_words_features_filtered(words):
    return {
        word:1 for word in words \
        if not word in useless_words}
```

```python
build_bag_of_words_features_filtered(text_words)
```

```python
filtered_words = [word for word in text_words if not word in useless_words]
type(filtered_words)
```

```python
len(filtered_words)
```

```python
from collections import Counter

word_counter = Counter(filtered_words)
```

```python
most_common_words = word_counter.most_common()[:10]
```

```python
most_common_words
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sorted_word_counts = sorted(list(word_counter.values()), reverse=True)
plt.style.use('fivethirtyeight')
#plt.figure(figsize=(14,7))
plt.loglog(sorted_word_counts)
plt.ylabel("Freq (log)")
plt.xlabel("Word Rank (log)");
```
